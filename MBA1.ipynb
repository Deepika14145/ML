```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtent.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('Order1.csv')
df.head(10)
```

```python
transactions = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).tolist()
```

```python
te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_data, columns = te.columns)
```

```python
frequent_items = apriori(df_encoded, min_support = 0.01, use_colnames = Ture)
```

```python
rules = association_rules(frequent_items, metric="life", min_threshold= 1.0)
```
